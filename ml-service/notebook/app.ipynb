```python
import pandas as pd
df=pd.read_csv("../data/dataset.csv");
df
```

```python
symptom_cols = [col for col in df.columns if "Symptom" in col]

all_symptoms = set()

for col in symptom_cols:
    all_symptoms.update(df[col].dropna().unique())

all_symptoms = sorted(all_symptoms)

print("Total symptoms:", len(all_symptoms))
```

```python
encoded_rows = []

for _, row in df.iterrows():

    symptoms_present = set(row[symptom_cols].dropna())

    encoded = {symptom: (symptom in symptoms_present) for symptom in all_symptoms}
    encoded["Disease"] = row["Disease"]

    encoded_rows.append(encoded)

encoded_df = pd.DataFrame(encoded_rows)
encoded_df.head()
```

```python
encoded_df = encoded_df.replace({True:1, False:0})
encoded_df.head()
```

```python
disease_to_specialist = {

# ---------------- Skin → Dermatologist ----------------
"Fungal infection":"Dermatologist",
"Allergy":"Dermatologist",
"Acne":"Dermatologist",
"Impetigo":"Dermatologist",
"Psoriasis":"Dermatologist",
"Drug Reaction":"Dermatologist",

# ---------------- Brain & Nerves → Neurologist ----------------
"Migraine":"Neurologist",
"Paralysis (brain hemorrhage)":"Neurologist",
"Cervical spondylosis":"Neurologist",

# ---------------- Stomach, Liver, Digestion → Gastroenterologist ----------------
"GERD":"Gastroenterologist",
"Peptic ulcer diseae":"Gastroenterologist",
"Gastroenteritis":"Gastroenterologist",
"Chronic cholestasis":"Gastroenterologist",
"Jaundice":"Gastroenterologist",
"Hepatitis A":"Gastroenterologist",
"Hepatitis B":"Gastroenterologist",
"Hepatitis C":"Gastroenterologist",
"Hepatitis D":"Gastroenterologist",
"Hepatitis E":"Gastroenterologist",
"Alcoholic hepatitis":"Gastroenterologist",

# ---------------- Women specific → Gynecologist ----------------
"Urinary tract infection":"Gynecologist",

# ---------------- Child dominant → Pediatricians ----------------
"Chicken pox":"Pediatricians",

# ---------------- Everything systemic → General physician ----------------
"Common Cold":"General physician",
"Flu":"General physician",
"Hypertension ":"General physician",
"Diabetes ":"General physician",
"Bronchial Asthma":"General physician",
"Pneumonia":"General physician",
"Malaria":"General physician",
"Dengue":"General physician",
"Typhoid":"General physician",
"Tuberculosis":"General physician",
"Hypothyroidism":"General physician",
"Hyperthyroidism":"General physician",
"Heart attack":"General physician",
"Varicose veins":"General physician",
"Osteoarthritis":"General physician",
"Arthritis":"General physician",
"Dimorphic hemmorhoids(piles)":"General physician"
}
```

```python
missing = set(encoded_df["Disease"]) - set(disease_to_specialist.keys())
print(missing)
print("Missing count:", len(missing))
```

```python
disease_to_specialist.update({

# liver infection → digestive system
"hepatitis A":"Gastroenterologist",

# joints / bones → first contact is GP (not orthopedician in your system)
"Osteoarthristis":"General physician",

# dizziness / balance disorder → neuro
"(vertigo) Paroymsal  Positional Vertigo":"Neurologist",

# low sugar emergency metabolic → GP
"Hypoglycemia":"General physician",

# immune disorder systemic → GP initial screening
"AIDS":"General physician"

})
```

```python
missing = set(encoded_df["Disease"]) - set(disease_to_specialist.keys())
print(missing)
print("Missing count:", len(missing))
```

```python
encoded_df.isnull().sum()
```

```python
encoded_df["specialist"] = encoded_df["Disease"].map(disease_to_specialist)
encoded_df[["Disease","specialist"]].head(10)
```

```python

X = encoded_df.drop(columns=["Disease","specialist"])
y = encoded_df["specialist"]

print(X.shape)
print(y.value_counts())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,     # keeps class balance
    random_state=42
)

print(X_train.shape, X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=500,
    class_weight="balanced",
    n_jobs=-1,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

```python
import joblib

artifact = {
    "model": model,
    "symptom_columns": list(X.columns),
    "classes": list(model.classes_)   # specialist labels
}

joblib.dump(artifact, "specialist_model.joblib")
```
